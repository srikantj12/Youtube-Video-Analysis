```
%sql
--top 10 channels
select channel_title
, sum(views) as total_views
, sum(likes) as total_likes 
, sum(dislikes) as total_dislikes 
, sum(likes) - sum(dislikes) as popularity
from cavideos_csv 
group by channel_title 
order by popularity desc
limit 10
```

```
%sql
--top 10 contraversal
select title
, sum(views) as total_views
, sum(likes) as total_likes 
, sum(dislikes) as total_dislikes 
, sum(dislikes)/sum(likes) as contraversal
from cavideos_csv 
group by title 
having total_dislikes > 100000
and total_likes > 100000
order by contraversal desc
limit 10
```

```
%sql
--viral_videos
select 
title
--,tags
,category_id
,count(*) as cnt_days_tranding
,sum(views) as total_views
,sum(likes) as total_likes
,sum(comment_count) as total_comments
from cavideos_csv 
where title is not null
group by 1,2--,3
having total_views is not null
order by cnt_days_tranding desc
limit 10
```

```
%sql
--viral_videos
select 
title
,tags
,category_id
,count(*) as cnt
,sum(views) as total_views
,sum(likes) as total_likes
,sum(comment_count) as total_comments
from cavideos_csv 
where title is not null
group by 1,2,3
having total_views is not null
order by cnt desc
limit 10
```
